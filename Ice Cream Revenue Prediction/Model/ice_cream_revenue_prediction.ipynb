```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('..\Dataset\IceCreamData.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Temperature'], df['Revenue'])
plt.show() 
```

```python
x, y = df[['Temperature']].values, df[['Revenue']].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=0)
```

```python
# By using linear regression
```

```python
linear = LinearRegression()
linear.fit(x_train, y_train)
```

```python
y_pred = linear.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color='green')
plt.plot(x_test, y_pred, color='red')
plt.show() 
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import max_error
from sklearn.metrics import r2_score

print("r2_Score: ",r2_score(y_pred, y_test))
```

```python
# By using polynomial regression
```

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x_train)
poly.fit(x_train, y_train)
```

```python
linear2 = LinearRegression()
linear2.fit(x_poly, y_train)
```

```python
y_pred = linear2.predict(poly.fit_transform(x_test))
```

```python
plt.scatter(x_test, y_test, color='green')
plt.plot(x_test, y_pred, color='red')
plt.show() 
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import max_error
from sklearn.metrics import r2_score

print("r2_Score: ",r2_score(y_pred, y_test))
```
